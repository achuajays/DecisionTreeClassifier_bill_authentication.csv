```python
import pandas as pd
import numpy as np
```

```python
n = pd.read_csv('/content/bill_authentication.csv')
```

```python
n.head()
```

```python
n.isna().sum()
```

```python
n['Class'].unique()
```

```python
X = n[['Variance',	'Skewness',	'Curtosis',	'Entropy']]
```

```python
y = n['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score
```

```python
x_t ,x_te , y_t , y_te = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
x_t
```

```python
x_te
```

```python
y_t
```

```python
y_te
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_t,y_t)
```

```python
p = model.predict(x_te)
```

```python
r2_score(p,y_te)
```

```python
X_new = [[1.185, -0.276, -0.148, 0.156]]
y_new = model.predict(X_new)
print(y_new)
```
